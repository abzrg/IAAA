# Diabetes Dataset

## Description

The Diabetes Dataset contains various health-related features of patients, such as glucose levels, insulin levels, BMI, age, etc., and a binary target variable indicating whether the patient has diabetes or not.

## Attributes

- __Pregnancies__: Number of times pregnant
- __Glucose__: Plasma glucose concentration after 2 hours in an oral glucose tolerance test
- __BloodPressure__: Diastolic blood pressure ($\textrm{mm Hg}$)
- __SkinThickness__: Triceps skin fold thickness ($\textrm{mm}$)
- __Insulin__: 2-Hour serum insulin ($\mu\text{U/ml}$) <- micro-unit per mililiter?
- __BMI__: Body mass index ($\frac{\text{weight}}{\text{height}^2}$)
- __DiabetesPedigreeFunction__: Diabetes pedigree function (a function which scores likelihood of diabetes based on family history)
- __Age__: Age of the patient (years)
- __Outcome__: Target variable (0 if non-diabetic, 1 if diabetic)

## Objective

The objective of this dataset is to perform a classification task to predict whether a patient has diabetes or not based on the given health-related features.

# Background

Diabetes is a chronic medical condition that occurs when the body is unable to properly regulate blood sugar (glucose) levels. This can lead to high blood sugar levels, which, if left untreated, can cause various complications. There are two main types of diabetes: Type 1 and Type 2.

**Type 1 Diabetes:**
In Type 1 diabetes, the immune system mistakenly attacks and destroys the insulin-producing cells in the pancreas. Insulin is a hormone that helps regulate blood sugar by allowing cells to absorb and use glucose for energy. Without insulin, glucose builds up in the bloodstream, leading to high blood sugar levels. Type 1 diabetes is usually diagnosed in children and young adults and requires lifelong insulin therapy.

**Type 2 Diabetes:**
Type 2 diabetes is characterized by the body's resistance to the effects of insulin or insufficient insulin production. Risk factors for Type 2 diabetes include _genetics_, _age_, _obesity_, _lack of physical activity_, and _poor diet_. In this type, the pancreas may produce insulin, but the body's cells do not respond effectively to it. Over time, this insulin resistance and inadequate insulin production result in elevated blood sugar levels.

---

**Glucose:**
Glucose is a simple sugar and a primary source of energy for the body's cells. It is a crucial component of carbohydrates, which are broken down during digestion into glucose. Foods like bread, rice, fruits, and starchy vegetables are common sources of glucose. Once absorbed into the bloodstream, glucose circulates and provides energy to cells for various physiological functions.

The body tightly regulates blood glucose levels to ensure a stable and optimal environment for cellular activities. Insulin, a hormone produced by the pancreas, plays a central role in this regulation by facilitating the uptake of glucose into cells. Elevated blood glucose levels, as seen in conditions like diabetes, can lead to various health issues and complications.

**Insulin:**
Insulin is a hormone produced by beta cells in the pancreas, an organ located behind the stomach. Its primary function is to regulate blood glucose levels. When you eat, especially foods containing carbohydrates, the digestive system breaks down these carbohydrates into glucose, which enters the bloodstream.

In response to rising blood glucose levels, the pancreas releases insulin. Insulin acts as a key that unlocks cells, allowing them to take in glucose from the bloodstream. Once inside the cells, glucose is used for energy production. Insulin also helps store excess glucose in the liver for later use when blood glucose levels decrease, such as between meals or during physical activity.

In conditions like Type 1 diabetes, the immune system mistakenly attacks and destroys the insulin-producing cells in the pancreas, leading to insufficient insulin production. In Type 2 diabetes, the body's cells become resistant to the effects of insulin, and the pancreas may not produce enough insulin to compensate, resulting in elevated blood glucose levels. Proper insulin function is crucial for maintaining blood glucose within a normal range and supporting overall health.

# Handling Anamolies

Now, Let's fix anamolies (zero values for skin thickness, blood pressure, and bmi)

```python
from sklearn.impute import SimpleImputer

# 1. Replace Zero Skin Thickness:
non_zero_skin_thickness_mean = df[df["SkinThickness"] > 0][
    "SkinThickness"
].mean()
df["SkinThickness"].replace(0, non_zero_skin_thickness_mean, inplace=True)

# 2. Replace Blood Pressure of 0:
# Calculate the mean blood pressure excluding zero values
non_zero_bp_mean = df[df["BloodPressure"] > 0]["BloodPressure"].mean()
df["BloodPressure"].replace(0, non_zero_bp_mean, inplace=True)

# 3. Replace Zero BMI Values:
# Calculate the mean BMI excluding zero values
non_zero_bmi_mean = df[df["BMI"] > 0]["BMI"].mean()
df["BMI"].replace(0, non_zero_bmi_mean, inplace=True)

# Handling missing values in other columns (if any)
# Use SimpleImputer to replace missing values with the mean or median
imputer = SimpleImputer(strategy="mean")  # One can use 'median' as well
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```

# Model Selection

```python
from sklearn. linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
# fmt: off
models = {
                       "Logistic Regression": LogisticRegression(),
                       "K-Nearest Neighbors": KNeighborsClassifier(),
                             "Decision Tree": DecisionTreeClassifier(),
    "Support Vector Machine (Linear Kernel)": LinearSVC(),
       "Support Vector Machine (RBF Kernel)": SVC(),
                            "Neural Network": MLPClassifier(),
                             "Random Forest": RandomForestClassifier(),
                         "Gradient Boosting": GradientBoostingClassifier(),
}
# fmt: on


for name, model in models.items():
    model.fit(X_train, y_train)
    print(name, "trained.")
```

# Model Evaluation

```python
for name, model in models.items():
    print(f"{name:>40}: {model.score(X_test, y_test) * 100}%")
```
